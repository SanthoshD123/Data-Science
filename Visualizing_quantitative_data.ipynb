```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
import numpy as np
```

```python
listings = pd.read_csv('listings.csv')
```

```python
listings
```

```python
plt.hist(listings['price'])
plt.xlabel('price(in US dollors)')
plt.show()
```

```python
plt.hist(listings['price'],bins=np.arange(0,1100,40))
plt.xlabel('price(in US dollors)')
plt.show()
```

```python
plt.scatter(x=listings['price'],y=listings['number_of_reviews'])
plt.xlabel('price')
plt.ylabel('number_of_reviews')
plt.show()
```

```python
plt.scatter(x=listings['price'],y=listings['number_of_reviews'],s=5)
plt.xlabel('price')
plt.ylabel('number_of_reviews')
plt.xlim(0,1100)
plt.show()
```
